```python
from requests_html import HTMLSession
```

```python
session = HTMLSession()
url = "http://www.math.keio.ac.jp/~ishikawa/indexe.html#HC"
r = session.get(url)
```

```python
xs = r.html.links
xs = [x for x in xs if 'http://www.math.keio.ac.jp/~ishikawa/QLEJ/index' in x ]
xs.sort()
xs
```

```python
xs = [
    f'http://www.math.keio.ac.jp/~ishikawa/QLEJ/indexj{i:03}.html'
    for i in range(1,125)]
xs+=["http://www.math.keio.ac.jp/~ishikawa/QLEJ/indexj085_1.html"]
xs.sort()
links = xs
links
```

```python
htmls = [
   session.get(link).html
   for link in links
]
```

```python
def get_title(html):
    y =  html.find('#header span h1')
    y = y if y else html.find('#header h1')
    return y[0].text
titles = [get_title(html) for html in htmls]
titles
```

```python
def edit_title(title):
    x, *xs = title.split('.')
    xs = f"{int(x):02}", *xs
    return '.'.join(xs)
titles = [edit_title(title) for title in titles]
titles
```
